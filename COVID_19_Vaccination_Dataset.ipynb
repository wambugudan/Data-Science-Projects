## Importing all the required Libraries

```python
#Importing all the required Libraries

import pandas as pd
import numpy as nm
from sklearn import preprocessing
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import Binarizer
from sklearn.decomposition import PCA
from numpy import set_printoptions
from IPython.display import display
```

## Task 1: Extarcting Data from the data source

```python
# Reading data from source file and printing it in the output

path = "C:\\pydatafile\\country_vaccinations.csv"
df = pd.read_csv(path)

print("\n Pringting raw data: \n\n", display(df))
```

## Task 2: Labelling Data

```python
# Performing data Labelling & pringting the output

Encoder = preprocessing.LabelEncoder()
Encoded_Dataframe = df.apply(Encoder.fit_transform)

print("\n Pringting Encoded data: \n", display(Encoded_Dataframe))
```

## Task 3: Data Splitting

```python
# Seperating Numeric Data from attributes labels

Numeric_Data = Encoded_Dataframe.values
print("\n Printing the data without attributes labels: \n\n", Numeric_Data)
```

```python
# Splitting Data
x_attributes = Numeric_Data[:,:-1]
y_attributes = Numeric_Data[:,-1]

print("\n Splitted Input Attributes sample: \n\n", x_attributes, "\n\n Splitted Output Attributes sample: \n", y_attributes)
```

## Task 4: Normalizing Data

```python
# Normalizing data and printing the output

normalized_x_attributes = preprocessing.normalize(x_attributes)
normalized_y_attributes = preprocessing.normalize(x_attributes)
print("\n Normalized Input Attributes: \n\n", normalized_x_attributes)
```

## Task 5: Standardizing Data

```python
# Standardizing Data & Printing the output

standardized_x_attributes = preprocessing.scale(x_attributes)
print("\n Standardized input attributes: \n\n", standardized_x_attributes)
```

## Task 6: Binarizing Data

```python
#Binarizing Data

Binarizer = Binarizer(threshold = 1.0).fit(x_attributes)
Binarized_x_attributes = Binarizer.transform(x_attributes)
print("\n Binarized output: \n\n", Binarized_x_attributes)
```

## Task 7: Compressing Data

```python
# Decomposing the number of attributes to two and printing the output

pca = PCA(n_components = 2)
pca.fit(x_attributes)
Decomposed_x_attributes = pca.transform(x_attributes)

print("\n Decomposed Input Attributes: \n\n", Decomposed_x_attributes)
```
