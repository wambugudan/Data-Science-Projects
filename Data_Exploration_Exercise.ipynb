```python
##Import data
# Data URL
url = "https://github.com/stedy/Machine-Learning-with-R-datasets/blob/master/credit.csv"

# Import urlreq lib
from urllib.request import urlretrieve

# Get & Save data
urlretrieve(url, "credit.csv")
```

```python
# Read Credit CSV
import pandas as pd
credit_df = pd.read_csv('insurance.csv')
```

```python
# Import visualization libraries
import plotly.express as px
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Configure Display aesthetics
sns.set_style('whitegrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (10, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
credit_df.age.describe()
```

```python
# Visualizing age using Histogram & box plot
fig = px.histogram(credit_df, x='age',
                   marginal ='box',
                   color = 'smoker',
                   color_discrete_sequence=['blue', 'orange'],
                   nbins = 47,
                   title = 'Age Distribution')
fig.update_layout(bargap = 0.1)
fig.show()
```

```python
# Visualizing BMI
fig = px.histogram(credit_df, x = 'bmi',
                  marginal='box',
                  color = 'sex',
                  color_discrete_sequence=['red', 'gray'],
                  title= 'BMI Distribution'
                  )

fig.update_layout(bargap=0.1)
fig.show()
```

```python
# Visualizing the Usage cost

fig = px.histogram(credit_df, x= 'charges', marginal='box',
                   color='smoker', color_discrete_sequence=['green', 'gray'],
                   title= 'Annual Medical Charges')

fig.update_layout(bargap = 0.1)
fig.show()
```

```python
# Visualizing Smoking and Religion
fig = px.histogram(credit_df, x = 'bmi',
                   color ='region', color_discrete_sequence=['orange', 'gray', 'blue', 'brown'],
                   title = 'Regions Distribution')
fig.update_layout(bargap = 00.1)
fig.show()
```

```python
# Visualizing Smokers
px.histogram(credit_df, x= 'smoker', color = 'sex', title = 'smoker')
```

```python
# Visualizing Sex
px.histogram(credit_df, x= 'region', color = 'sex', title = 'Region')
```

```python
# Visualizing Relationships between Age & Charges

fig = px.scatter(credit_df, x = 'age', y = 'charges',
                 color = 'smoker', opacity = 0.95, hover_data = ['sex'],
                 title = 'Age Vs Charges')
fig.update_traces(marker_size = 5)
fig.show()
```
