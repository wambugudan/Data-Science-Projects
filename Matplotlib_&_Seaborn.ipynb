```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
```

```python
# Data to be plotted

years = range(2000, 2012)
apples = ([0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939])
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]

print(type(apples))
```

```python
# Plotting

plt.plot(years, apples)
plt.plot(years, oranges);
```

```python
# Adding line decorations
#

plt.plot(years, apples, 'x-b')
plt.plot(years, oranges, 'o-r');
```

```python
# Adding labels

plt.plot(years, apples, 'x-b')
plt.plot(years, oranges, 'o-r')

plt.xlabel('Years')
plt.ylabel('Yield (Tonnes per acre)')

plt.title('Fruits yield in Nyeri')
plt.legend('Apples', 'Oranges');
```

```python
# Improving Chart Appearance using seaborn

sns.set_style("whitegrid")

# Adding labels

plt.plot(years, apples, 'x-b')
plt.plot(years, oranges, 'o-r')

plt.xlabel('Years')
plt.ylabel('Yield (Tonnes per acre)')

plt.title('Fruits yield in Nyeri')
plt.legend('Apples', 'Oranges');
```

```python
# Improving Chart Appearance using seaborn

sns.set_style("darkgrid")

# Adding labels

plt.plot(years, apples, 'x-b')
plt.plot(years, oranges, 'o-r')

plt.xlabel('Years')
plt.ylabel('Yield (Tonnes per acre)')

plt.title('Fruits yield in Nyeri')
plt.legend('Apples', 'Oranges')

# Changing the chart size
plt.figure(figsize = (20,7));
```

```python
# Importing seaborn data

flw_df = pd.read_csv('iris.csv')

flw_df
```

```python
# Get a list of species in iris dataframe

species = flw_df.species.unique()

species
```

```python
# Plot sepal dimensions using matplotlib

plt.plot(flw_df.sepal_length, flw_df.sepal_width)
```

```python
# Plot sepal dimensions using seaborn

sns.scatterplot('sepal_length', 'sepal_width', hue = 'species', s = 100, data = flw_df);
```

```python
# Plotting a histgram
# Creating sub dataframes
setosa_df = flw_df[flw_df.species == 'setosa']
versicolor_df = flw_df[flw_df.species == 'versicolor']
virginica_df = flw_df[flw_df.species == 'virginica']
```

```python
# Ploting
import numpy as np

bin_size = np.arange(2, 5, 0.25)

plt.hist(flw_df.sepal_width, bins=bin_size)
```

```python
# Combined hists

plt.hist(setosa_df.sepal_width, alpha = 0.99, bins = bin_size)
plt.hist(versicolor_df.sepal_width, alpha = 0.4, bins = bin_size)
```

```python
# Stacked hist

sns.set_style('darkgrid')
plt.hist([setosa_df.sepal_width, virginica_df.sepal_width, versicolor_df.sepal_width],  bins=bin_size, stacked = True);
```

```python
# Bar combine with line charts
plt.bar(years, oranges)
plt.plot(years, oranges, 'o--r')
```

```python
# Stacked Bar Chart

plt.bar(years, oranges)
plt.bar(years, apples, bottom=apples)
plt.title('Fruits yield')
```

```python
# Using averages in bar chart

tips_df = sns.load_dataset("tips")

tips_df
```

```python
# print averages Manually

bill = tips_df.groupby('day')[['total_bill']].mean()

bill
```

```python
plt.bar(bill.index, bill.total_bill)
```

```python
# Plot averahes using seaborn

sns.barplot('day', 'total_bill', hue = 'sex', data=tips_df);
```

```python
# Plot averahes using seaborn

sns.barplot('total_bill', 'day', hue = 'sex', data=tips_df);
```

```python
# Heat maps

# Load flights data

flights_df = sns.load_dataset('flights')

flights_df
```

```python
# Convert dataframe to 2-d array

flights = flights_df.pivot('month', 'year', 'passengers')

flights
```

```python
# Plot heat map

sns.heatmap(flights, fmt="", annot= True, cmap='Reds')
```

```python
# Creating multiple charts on the same grid

fig, axes = plt.subplots(2,2, figsize=(16,8))

#fig, axes = plt.subplots(2, 3, figsize=(16, 8))

axes[0,0].plot(years,apples, 's-b')
axes[0,0].plot(years,oranges, 'o--r')
axes[0,0].set_xlabel('Years')
axes[0,0].set_ylabel('Yield (tons/ha)')
axes[0,0].set_title('Crop Yields')

#plt.tight_layout(pad=2)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 8))

# Use the axes for plotting
axes[0,0].plot(years, apples, 's-b')
axes[0,0].plot(years, oranges, 'o--r')
axes[0,0].set_xlabel('Year')
axes[0,0].set_ylabel('Yield (tons per hectare)')
axes[0,0].legend(['Apples', 'Oranges']);
axes[0,0].set_title('Crop Yields in Kanto')


# Pass the axes into seaborn
axes[0,1].set_title('Sepal Length vs. Sepal Width')
sns.scatterplot(x=flw_df.sepal_length,
                y=flw_df.sepal_width,
                hue=flw_df.species,
                s=100,
                ax=axes[0,1]);

# Use the axes for plotting
axes[0,2].set_title('Distribution of Sepal Width')
axes[0,2].hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width],
         bins=np.arange(2, 5, 0.25),
         stacked=True);

axes[0,2].legend(['Setosa', 'Versicolor', 'Virginica']);

# Pass the axes into seaborn
axes[1,0].set_title('Restaurant bills')
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df, ax=axes[1,0]);

# Pass the axes into seaborn
axes[1,1].set_title('Flight traffic')
sns.heatmap(flights_df, cmap='Blues', ax=axes[1,1]);

# Plot an image using the axes
axes[1,2].set_title('Data Science Meme')
axes[1,2].imshow(img)
axes[1,2].grid(False)
axes[1,2].set_xticks([])
axes[1,2].set_yticks([])

plt.tight_layout(pad=2);
```
